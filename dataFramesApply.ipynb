```python
def salary_category(salary):
    if salary < 5000:
        return "low"
    elif 50000 <= salary <80000:
        return "medium"
    else:
        return "high"
```

```python
df['Salary'].apply(salary_category)
```

```python
df['Salary_Category'] = df['Salary'].apply(salary_category)
```

```python
df
```

```python
def performance_score_category(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df['Adjust'] = df['Experience'].apply(performance_score_category)
```

```python
df
```

```python
df['New_Performance_Score'] = df['Performance_Score'] + df['Adjust']
```

```python
df
```

```python
def adjust_new(row):
    if row['Experience'] > 10:
        return row['Performance_Score'] + 1
    else:
        return row['Performance_Score']
```

```python
df["Adjust_Score"] = df.apply(adjust_new, axis=1)
```

```python
df
```

```python
df['Formatted_Name'] = df['Name'].apply(lambda x: x.replace('_', ' '))
```

```python
df
```

```python
df.drop(['Adjust'], axis=1, inplace=True)
```

```python
df.drop(['Adjust_Score'], axis=1, inplace=True)
```
