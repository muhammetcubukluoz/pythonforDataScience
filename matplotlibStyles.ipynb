```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, "g")
my_axes.plot(data2, data1, color="blue", alpha=0.5) #color=hex color
plt.show()
```

```python
(new_fix, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="blue", linewidth=3)
new_axes.plot(data1, data1 + 3, color="green", linewidth=2)

new_axes.plot(data1, data1 + 4, color="green", linestyle="-.")
new_axes.plot(data1, data1 + 5, color="green", linestyle=":")
new_axes.plot(data1, data1 + 6, color="green", linestyle="--")

new_axes.plot(data1, data1 + 7, color="yellow", linestyle="--", marker="o", markersize=8, markerfacecolor="red")
new_axes.plot(data1, data1 + 8, color="yellow", linestyle="--", marker="o", markersize=4)
```
