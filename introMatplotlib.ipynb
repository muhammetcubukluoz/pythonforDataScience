```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list, weight_list, "b")
#plt.savefig("test.png")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
#numpy
```

```python
np_age_list = np.array(age_list)
```

```python
np_age_list
```

```python
np_weight_list = np.array(weight_list)
```

```python
np_weight_list
```

```python
plt.plot(np_age_list, np_weight_list, "b")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20) # even number same range
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1, numpy_arr2, "b")
plt.show()
```

```python
plt.plot(numpy_arr1, numpy_arr2, "g--")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1, numpy_arr2, "g--")
plt.subplot(1,2,2)
plt.plot(numpy_arr2, numpy_arr1, "g*-")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.6,0.6]) #[left,bottom,weight size,height size]
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("X Axis vs Y Axis")
plt.show()
```

```python
new_fig = plt.figure(dpi=150)

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1, numpy_arr1**2, label ="numpy array **2")
new_axes.plot(numpy_arr1, numpy_arr1**3, label ="numpy array **3")
new_axes.legend(loc="best")
plt.show()
```
