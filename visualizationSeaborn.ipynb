```python
plt.scatter("Height", "Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.scatterplot(x="Height", y="Weight", data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.scatterplot(x="Height", y="Weight", hue="Sex", data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.scatterplot(x="Height", y="Weight", hue="Medal", data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
#sns.set_theme(rc={"figure.dpi":150})
sns.scatterplot(x="Height", y="Weight", hue="Sex", style="Medal", size="Age", data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height with Age")
plt.show()
```

```python
sns.reset_defaults()

sns.set_style("whitegrid")
sns.lineplot(x="Height", y="Weight",data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="Height", y="Weight", hue="Sex",data=data)
plt.xlabel("Height of Athlete")
plt.ylabel("Weight of Athlete")
plt.title("Athelete Weight vs Height with Sex")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", data=data)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height", hue="Sex", kind="kde", data=data)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x="Medal", y="Height", hue="Sex", data=data)
plt.title("Medal vs Height")
plt.show()
```

```python
data["Season"].unique()
```

```python
sns.catplot(x="Medal", y="Height", hue="Sex", col="Season", data=data)
plt.title("Medal vs Height with Season and Sex")
plt.show()
```
