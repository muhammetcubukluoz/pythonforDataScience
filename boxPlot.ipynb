```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# median, quartile(25%, 50%, 75%), min max 1,5 * IQR, outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(4,8))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
#plt.grid(True)
plt.show()
```

```python

data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(4,8))
sns.boxplot(y=data)

for i, val in enumerate(data):
    plt.text(
        x=0,            # boxplot single but x=0
        y=val,
        s=str(val),
        ha='left',
        va='center',
        fontsize=8,
        color="green"
    )

plt.title("Box Plot with Values")
plt.ylabel("Data Value")
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(13,8))

plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```
