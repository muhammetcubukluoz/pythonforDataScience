```python
data["quality"].value_counts().plot(kind="bar")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.title("Quality Count Bar Chart")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.scatterplot(x="fixed acidity", y="density", hue="quality",data=data)
plt.xlabel("Acidity of Wine")
plt.ylabel("Density of Wine")
plt.title("Wine Acidity vs Density")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.scatterplot(x="fixed acidity", y="citric acid", hue="quality",data=data)
plt.xlabel("Acidity of Wine")
plt.ylabel("Citric Acid of Wine")
plt.title("Wine Acidity vs Citric Acid")
plt.show()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.lineplot(x="alcohol", y="quality",data=data)
plt.xlabel("Alcohol of Wine")
plt.ylabel("Quality of Wine")
plt.title("Wine Alcohol vs Quality")
```

```python
sns.set_style("whitegrid")
sns.boxplot(x="quality", y="alcohol",data=data)
plt.xlabel("Quality of Wine")
plt.ylabel("Alcohol of Wine")
plt.title("Wine Alcohol vs Quality")
```

```python
data["pH"].unique()
```

```python
sns.set_style("whitegrid") #darkgrid, dark,white
sns.lineplot(x="pH", y="quality",data=data)
plt.xlabel("PH of Wine")
plt.ylabel("Quality of Wine")
plt.title("Wine PH vs Quality")
```

```python
plt.figure(figsize=(4,8))
sns.boxplot(y="pH", x="quality", data=data)
plt.title("pH vs quality")
plt.ylabel("pH of Wine")
plt.grid(True)
plt.show()
```

```python
sns.catplot(x="quality", y="sulphates", data=data)
plt.title("Medal vs Height with Season and Sex")
plt.show()
```

```python
sns.scatterplot(data=data, x="free sulfur dioxide", y="total sulfur dioxide", hue="quality", alpha=0.6)
plt.xlabel("free sulfur dioxide")
plt.ylabel("total sulfur dioxide")
plt.title("Free vs Total Sulfur Dioxide")
plt.show()
```

```python
sns.scatterplot(x="alcohol", y="sulphates", hue="quality", data=data, palette="viridis")
plt.xlabel("Alcohol of Wine")
plt.ylabel("Sulfur Dioxide")
plt.title("Alcohol vs Sulphates by Quality")
plt.show()
```

```python
sns.histplot(data=data, x="alcohol", hue="quality", kde=True, element="step",  palette="viridis") #poly,bars
plt.title("Alcohol Distribution by Quality")
plt.show()
```

```python
sns.pairplot(data, vars=["alcohol", "sulphates", "citric acid", "density", "pH"], hue="quality")
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")

columns = [col for col in data.columns if col != "Id"]
(fig,ax) = plt.subplots(4,4, figsize=(15,15))
ax = ax.flatten()

for i,column in enumerate(columns):
    sns.kdeplot(
        data=data,
        x=column,
        hue=data.quality,
        ax=ax[i],
    )

    ax[i].set_title(f"{column} Distribution")


for i in range(i+1, len(ax)):
    ax[i].axis("off")

plt.tight_layout()
plt.show()
```
